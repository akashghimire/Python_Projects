```python
import requests
import bs4
```

```python
result = requests.get("https://en.wikipedia.org/wiki/Jonas_Salk")
type(result)
```

```python
result.text
```

```python
soup = bs4.BeautifulSoup(result.text,"lxml")
```

```python
soup
```

```python
soup.select('p')
```

```python
soup.select("title")[0].getText()
```

```python
first = soup.select(".toctext")[0]
```
